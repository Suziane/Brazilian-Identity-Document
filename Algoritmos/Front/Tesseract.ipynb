```python
import os
import cv2
import pytesseract
import matplotlib.pyplot as plt

base_folder = '/content/drive/MyDrive/imagens_cortadas'
image_list_file = '/content/drive/MyDrive/dataset/test_files.txt'

config = r'--oem 3 --psm 7'

def perform_ocr(image_path):
    img = cv2.imread(image_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    text = pytesseract.image_to_string(img, lang="eng", config=config)
    text = text.replace("\n", "").replace("\x0c", "").replace("\u2014", "")
    return text, img

# Inicializando o processo de OCR

with open(image_list_file, 'r') as f:
    folders_to_process = f.read().strip().split(',')
    folders_to_process = [arquivo.strip() for arquivo in folders_to_process]

for folder_name in os.listdir(base_folder):
    if folder_name in folders_to_process:
        folder_path = os.path.join(base_folder, folder_name)
        if os.path.isdir(folder_path):
            output_file = f'/content/drive/MyDrive/Tesseract/resultados_eng/{folder_name}/hipotese.txt'
            output_folder = os.path.dirname(output_file)

            if not os.path.exists(output_folder):
                os.makedirs(output_folder)

            ocr_results = []

            for filename in os.listdir(folder_path):
                if filename.endswith(('.png', '.jpg', '.jpeg')):
                    image_path = os.path.join(folder_path, filename)
                    ocr_text, image = perform_ocr(image_path)
                    ocr_results.append((filename.split('.')[0], ocr_text))
                    #print(f'OCR para {filename}: {ocr_text}')

                    # Plotando a imagem com o texto OCR dentro do loop
                    # plt.figure(figsize=(8, 6))
                    # plt.imshow(image)
                    # plt.title(f'OCR para {filename}: {ocr_text}', fontsize=10)
                    # plt.axis('off')
                    # plt.show()

            with open(output_file, 'w') as txt_file:
                for image_id, ocr_text in ocr_results:
                    txt_file.write(f'{image_id} "{ocr_text}"\n')

            print(f'Todos os resultados foram salvos em {output_file}')
```
